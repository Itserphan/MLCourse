```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3]
y = [4,5,6]

def de_mean(values):
    values_mean = np.mean(values)
    return [xi - values_mean for xi in values]

def covariance(a,b):
    n = len(a)
    return np.dot(de_mean(a),de_mean(b)) / (n-1)
#np.dot baraye zarb chand List az ham estefade mishavad
    
    
# print(covariance(x,y))   

page_requests = np.random.normal(50.0, 10.0, 1000)


page_speed = np.random.normal(3.0, 1.0, 1000)
page_visits = np.random.normal(50.0, 10.0, 1000) / page_speed

plt.scatter(page_speed, page_visits)

print(covariance(page_speed,page_visits))
```

```python
def correlation(x,y):
    std_x = x.std()
    std_y = y.std()
    return covariance(x,y) / std_x / std_y
print(correlation(page_speed, page_visits))
#adad bargardonde shode beine 1 va -1 mibashad
```
