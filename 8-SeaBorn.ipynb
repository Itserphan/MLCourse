```python
%matplotlib inline
import pandas as pd

data = pd.read_csv("cars.csv")
gear_counts = data['# Gears'].value_counts()
gear_counts.plot(kind = "bar")

```

```python
import seaborn as sb
#convert seaborn keys to matplotlib keys and combining them.
sb.set()
gear_counts.plot(kind = "bar")
```

```python
#nemoodar noghtei + nemoodar khati
sb.lmplot(x="Cylinders",y="CombMPG", data=data)
```

```python
#Nemoodar jaabeyi baraye tamami mavared 
sb.set(rc={'figure.figsize':(15,5)})
ax = sb.boxplot(x="Mfr Name",y="CombMPG", data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```

```python
#nemoodar jaabeyi + faravani tajamoii
sb.set(rc={'figure.figsize':(15,5)})
ax = sb.swarmplot(x="Mfr Name",y="CombMPG", data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```

```python
#shemordan teedad
sb.set(rc={'figure.figsize':(15,5)})
ax = sb.countplot(x="Mfr Name", data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```
