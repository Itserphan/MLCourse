```python
import statsmodels.api as sm 
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = result[['CombMPG', 'Cylinders', 'Eng Displ']]
y = result[['CombMPG']]
X[['CombMPG', 'Cylinders', 'Eng Displ']] = scale.fit_transform(X[['CombMPG', 'Cylinders', 'Eng Displ']])
est = sm.OLS(y, X).fit()
est.summary()
```
