```python
#using numpy to get mean of 20000 random numbers
import numpy as np
numbers = np.random.normal(25000, 15000, 20000)
np.mean(numbers)
```

```python
#matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(numbers, 200)
plt.show()
```

```python
numbers = np.append(numbers, [100000000])
```

```python
#mean
np.mean(numbers)
```
