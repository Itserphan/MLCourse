```python
import pandas as pd
import numpy as np
r_cols = ['User_ID', 'Movie_ID', 'Rating']
ratings = pd.read_csv('data/movies_data.txt',sep='\t', names=r_cols, usecols= range(3))

m_cols = ['Movie_ID', 'Title']
movies = pd.read_csv('data/movies_item.txt', sep='|', names = m_cols, usecols= range(2))
```

```python
ratings = pd.merge(movies, ratings)
ratings.head(10)
```

```python
movieRatings = ratings.pivot_table(index='User_ID', columns='Title', values='Rating')
movieRatings.head(10)
#چک میکنه که هر کاربر کدام فیلم هارا دیده و به آنها رای داده
```

```python
StarWarsRating = movieRatings['Star Wars (1977)']
print(StarWarsRating)
```

```python
SimilarMovies = movieRatings.corrwith(StarWarsRating)
SimilarMovies = SimilarMovies.dropna()
df = pd.DataFrame(SimilarMovies)
df.head()
```

```python
SimilarMovies.sort_values(ascending=False)
```

```python
MovieStats = ratings.groupby('Title').agg({'Rating':[np.size, np.mean]})
MovieStats.head(10)
#اسم فیلم با تعداد رای ها و میانگین رای ها
```

```python
PopularMovies = MovieStats['Rating']['size'] >= 100
MovieStats[PopularMovies].sort_values([('Rating', 'mean')], ascending = False)[:15]
```

```python
df2 = MovieStats[PopularMovies].join(pd.DataFrame(SimilarMovies, columns=['Similarity']))
df2.head()
```
