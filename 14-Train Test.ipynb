```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(3.0, 1.0, 100) / pageSpeeds

plt.scatter(pageSpeeds, purchaseAmount)
```

```python
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

```python
plt.scatter(trainX, trainY, color = 'r')
plt.scatter(testX, testY , c = 'b')
```

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x, y, 8))
print(p4)
```

```python
xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0, 7])
axes.set_ylim([0, 200])
plt.scatter(x,y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
test_x = np.array(testX)
test_y = np.array(testY)

plt.scatter(test_x, test_y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(test_y, p4(test_x ))
print(r2)
```

```python
r2_train = r2_score(trainY, p4(trainX))
print(r2_train)
```
